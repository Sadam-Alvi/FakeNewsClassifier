```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

def preprocess_alphanumeric(text):
    if not isinstance(text, str):
        return text, []
    
    # Keep only letters, numbers, and spaces
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    # Lowercase
    text = text.lower()
    
    # Tokenization
    tokens = word_tokenize(text)
    
    # Stop word removal
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    cleaned_text = ' '.join(tokens)
    
    return cleaned_text, tokens

# Test it
sample_text = """Hello World! This is a test. Price: $99.99. 
Contact: user@example.com. Call 123-456-7890 today!!! #NLP #AI"""

cleaned_text, tokens = preprocess_alphanumeric(sample_text)

print("Cleaned Text:", cleaned_text)
print("\nTokens:", tokens)
df
```

## Perfoming N-Gram
What is N-gram Analysis?
An N-gram is a sequence of N consecutive words.

Bigram = 2 consecutive words (e.g., "climate change", "fake news")
Trigram = 3 consecutive words (e.g., "breaking news today", "president donald trump")

Why is it useful for Fake News Detection?
Many real and fake news articles use different common phrases. Comparing bigrams and trigrams between Real (Label 0) and Fake (Label 1) h

```python

from sklearn.feature_extraction.text import CountVectorizer


def get_top_ngrams(corpus, n=2, top_n=20):
    """Extract top N-grams"""
    vec = CountVectorizer(ngram_range=(n, n), 
                          stop_words='english',
                          min_df=2)  # Ignore very rare n-grams
    bag_of_words = vec.fit_transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    return sorted(words_freq, key=lambda x: x[1], reverse=True)[:top_n]

# ====================== BIGRAMS ======================
print("=== TOP 20 BIGRAMS ===\n")

print("Label 0 - Real News:")
real_bigrams = get_top_ngrams(df[df['label'] == 0]['tags'], n=2, top_n=20)
for word, freq in real_bigrams:
    print(f"{word:40} : {freq:,}")

print("\nLabel 1 - Fake News:")
fake_bigrams = get_top_ngrams(df[df['label'] == 1]['tags'], n=2, top_n=20)
for word, freq in fake_bigrams:
    print(f"{word:40} : {freq:,}")

# ====================== TRIGRAMS ======================
print("\n\n=== TOP 15 TRIGRAMS ===\n")

print("Label 0 - Real News:")
real_trigrams = get_top_ngrams(df[df['label'] == 0]['tags'], n=3, top_n=15)
for word, freq in real_trigrams:
    print(f"{word:50} : {freq:,}")

print("\nLabel 1 - Fake News:")
fake_trigrams = get_top_ngrams(df[df['label'] == 1]['tags'], n=3, top_n=15)
for word, freq in fake_trigrams:
    print(f"{word:50} : {freq:,}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_top_ngrams(ngrams_list, title):
    words, freqs = zip(*ngrams_list)
    plt.figure(figsize=(10, 8))
    sns.barplot(x=list(freqs), y=list(words), palette='viridis')
    plt.title(title, fontsize=14)
    plt.xlabel('Frequency')
    plt.ylabel('Bigrams')
    plt.tight_layout()
    plt.show()

# ====================== PLOTS ======================

# For Real News (Label 0)
plot_top_ngrams(real_bigrams, "Top 20 Bigrams - Real News (Label = 0)")

# For Fake News (Label 1)
plot_top_ngrams(fake_bigrams, "Top 20 Bigrams - Fake News (Label = 1)")
```
